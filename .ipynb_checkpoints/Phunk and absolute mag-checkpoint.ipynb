```python
import lmfit

def eval_fit(phase, H, G):
    """Evaluation function for fitting. Required for lmfit."""
    return phot.HG.evaluate(np.radians(phase), H, G)

def fit(phase, mag, band = None, weights=None):
    """Fit a phase curve using the HG model."""
    model = lmfit.Model(eval_fit)
    params = lmfit.Parameters()

    params.add("H", value=15, min=0, max=30)
    params.add("G", value=0.15, min=0, max=1.0)

    phase = phase
    mag = mag
    weights_band = None

    # And fit
    result = model.fit(
        mag,
        params,
        phase=phase,
        method="least_squares",
        fit_kws={"loss": "soft_l1"},
    )
    return result
```

## Asteroid epoch to phase

```python
from astropy.time import Time
def date_str_to_jd(date_str):
    year_str, month_str, day_fraction_str = date_str.split()
    year = int(year_str)
    month = int(month_str)
    day_fraction = float(day_fraction_str)
    day_int = int(day_fraction)
    day_frac = day_fraction - day_int
    total_seconds = day_frac * 86400
    hours = int(total_seconds // 3600)
    minutes = int((total_seconds % 3600) // 60)
    seconds = total_seconds % 60
    iso_date_str = f'{year}-{month:02d}-{day_int:02d}T{hours:02d}:{minutes:02d}:{seconds:06.3f}'
    t = Time(iso_date_str, format='isot', scale='utc')
    return t.jd
```

```python
from astroquery.imcce import Miriade
epochs = ["2021 12 03.072179", "2021 12 03.877735", "2021 12 04.134679", 
         "2021 12 04.322179", "2021 12 04.579123", "2021 12 04.662456",
         "2021 12 04.974956", "2021 12 05.037456", "2021 12 05.405512", 
         "2021 12 05.738845", "2021 12 06.120789", "2021 12 06.384678", 
          "2021 12 06.627733", "2021 12 07.058289", "2021 12 07.398566",
          "2021 12 07.843011", "2021 12 08.218010", "2021 12 08.766621",
          "2021 12 09.218010", "2021 12 10.238843"
         ]
mags = [9.8, 9.9, 10.0, 10.1, 9.8, 10.1, 10.1, 10.2, 10.1, 10.1, 9.9, 10.2, 9.9, 10.2,
       10.3, 10.0, 10.4, 10.3, 10.2, 10.5]

epochs_jd = []
alphas = []
mags_list = []
m_red = []
for epoch in epochs:
    epoch_jd = date_str_to_jd(epoch)
    epochs_jd.append(epoch_jd)
    Mass_params = Miriade.get_ephemerides('Massalia', epoch=epoch_jd)
    alpha = float(Mass_params["alpha"])
    mags_val = float(Mass_params["V"])
    alphas.append(alpha)
    mags_list.append(mags_val)
    
    mag_red = mags_val - 5*np.log10(float(Mass_params["r"])*float(Mass_params["delta"]))
    m_red.append(mag_red)
```

```python
band = ["V"]*len(alphas)
```

```python
pc = phunk.PhaseCurve(mag=mags_list, epoch=epochs_jd, phase = alphas,  band=band, target = "Massalia")
```

```python
def hg_phase_function(H, G, alpha):
    """
    Calculate the apparent magnitude V of an asteroid at phase angle alpha
    using the HG phase function.

    Parameters:
    - H (float): Absolute magnitude of the asteroid.
    - G (float): Slope parameter of the asteroid (0  G  1).
    - alpha (float or array-like): Phase angle(s) in degrees.

    Returns:
    - V (float or array-like): Apparent magnitude(s) at phase angle(s) alpha.
    """
    # Convert alpha to radians
    alpha_rad = np.radians(alpha)

    # Compute Phi1 and Phi2
    phi1 = np.exp(-3.33 * np.power(np.tan(alpha_rad / 2), 0.63))
    phi2 = np.exp(-1.87 * np.power(np.tan(alpha_rad / 2), 1.22))

    # Compute the magnitude V(alpha)
    V = H - 2.5 * np.log10((1 - G) * phi1 + G * phi2)

    return V
```

```python
ph_an = np.linspace(0, max(ph)+2, 100)
mag_analy = hg_phase_function(13.8692477, 0.05871468, ph_an)

mag_MPC = hg_phase_function(13.93, 0.15, ph_an)
```

```python
plt.scatter(ph, real_mag, s = 7, c= "black", marker = "o", label = "data")
plt.plot(ph_an, mag_analy, c = "red",label =  "best fit HG")
plt.plot(ph_an, mag_MPC, c = "blue", label = "MPC values")
plt.legend()
plt.xlim(0, max(ph)+2)
plt.xlabel("Phase")
plt.ylabel("Reduced V")
# Get the current Axes instance
ax = plt.gca()

# Invert the x-axis and y-axis
ax.invert_yaxis()
```
